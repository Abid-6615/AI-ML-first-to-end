## Set

```python
a = []
print(type(a))

b = ()
print(type(b))

c= set()
print(type(c))

s = {1, 3, 5, 3, 7}

if (3 in s):
    print("Present")
else:
    print("Not Present")

sum = 0
for element in s:
    sum += element
print(sum)
```

```python
s = {4, 2, 8, 1, 3, 7, 5,}

s.add(10)
s.add(0)
print(s)

s.pop()
s.pop()
print(s)

s.remove(4)
s.remove(8)
print(s)
```

### Mathod of set

```python
a = {1, 2, 3}
b = {1, 2, 3, 4, 5, 6}

uni = a.union(b)
print(uni)

intr = a.intersection(b)
print(intr)

disjoint = a.isdisjoint(b)
print(disjoint)

sub = a.issubset(b)
print(sub)
```

## Dictionaries

```python
dic = {"name": "Abid", "age": 23, "address": "Uttara, Dhaka"}

print(dic)

print(dic["name"])

print(dic.get("address"))

dic["name"] = "Nishan"
print(dic)

dic.update({"year": "3rd year"})
print(dic)

del dic["year"]
print(dic)
```

```python
dic = {"name": "Abid", "age": 23, "address": "Uttara, Dhaka", "number": [10, 20, 30]}

keys = dic.keys()
print(keys)

dic["math"] = 60
print(dic)

values = dic.values()
print(values)

items = dic.items()
print(items)
```

```python
for key, value in dic.items():
    print(f"{key} --> {value}")
```

### Dictionaries Comprehension

```python
square = {x: x**2 for x in range(1, 11) if x%2 == 0}
print(square)
```

```python
Co_ordinates = [(10, 10.5), (20.5, 192), (101, 102)]
Locations = ["dhaka", "chattogram", "sylhet"]

exact_location = {co_or: loc for co_or, loc in zip(Co_ordinates, Locations)}

print(exact_location)
```

```python
lst = [10, 20, 10, 30, 30, 50, 30, 10, 20, 10, 10]

unique = set(lst)
lst = list(unique)
print(lst)
```
