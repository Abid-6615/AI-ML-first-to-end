# Problem 1

```python
#                ------> Smart Light Pilot <------
# Your lab is prototyping a Smart Light that turns on only when the room is bright enough.
# You are given two real numbers: the current brightness and a threshold.
# If brightness >= threshold. print ON; otherwise print OFF.

brightness, threshold = map(int, input().split())

if brightness >= threshold:
    print("ON")
else:
    print("OFF")
```

# Problem 2

```python
#               -----> Sensor Calibrator <-----
# You're calibrating a tiny sensor before training a model. Given x, min_v, and max_v, compute:
# norm = (x - min_v) / (max_v - min_v)
# Then print norm with 2 decimal places.
x, min_v, max_v = map(float, input().split())

norm = (x - min_v) / (max_v - min_v)

print(f"{norm:.2f}")
```
