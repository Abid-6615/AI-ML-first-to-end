```python
inp = input()
numbers = inp.split()

x = int(numbers[0])
y = int(numbers[1])

x_last = x % 10
y_last = y % 10
sum = x_last + y_last

print ("X value: ", x, "\nY value: ", y)
print ("The sum of the last digits is: ", sum)
```
