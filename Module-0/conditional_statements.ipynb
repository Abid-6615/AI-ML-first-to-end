## Loop Conditions

```python
for i in range(1, 11):
    print(i,"->", "Hello World!")
```

```python
for i in range(1, 11):
    for j in range(1, 4):
        print(i, "x", j, "=", i*j)
```

```python
count = 1
while count <= 10:
    print(count, "->", "Hello World!")
    count += 1
```

```python
sum = 0
for i in range(1, 11):
    print(i, "is prosessing...")
    if i % 2 == 0:
        continue
    sum += i
    print(i, "is added, current sum is", sum)
```

### Max and Min Problem

```python
inp = input()
numbers = inp.split()
x = int(numbers[0])
y = int(numbers[1])
z = int(numbers[2])

print("Input:\n", x, y, z)

print("\nOutput:")

if x > y and x > z:
    print("Largest number is = ", x)
    if y > z:
        print("Smallest number is = ", z)
    else:
        print("Smallest number is = ", y)
elif y > x and y > z:
    print("Largest number is = ", y)
    if x > z:
        print("Smallest number is = ", z)
    else:
        print("Smallest number is = ", x)
else:
    print("Largest number is = ", z)
    if x > y:
        print("Smallest number is = ", y)
    else:
        print("Smallest number is = ", x)
```

### Even, Odd, Positive, Negetive Problem

```python
num = int(input())
inp = input()
numbers = inp.split()

print("Input:", num, "\n", numbers)

even = 0
odd = 0
positive = 0
negative = 0

for i in range(num):
    x = int(numbers[i])

    if x > 0:
        positive += 1
    elif x < 0:
        negative += 1

    if x % 2 == 0:
        even += 1
    else:
        odd += 1

print("\nOutput:\n", "Even: ", even, "\n Odd: ", odd, "\n Positive: ", positive, "\n Negative: ", negative)
```
