# ploting 2d line plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hours = [2, 3, 4, 1, 5, 7, 3]
days = [1, 2, 3, 4, 5, 6, 7]

plt.plot(days, hours)
```

# 2d line plot from pandas

```python
df = pd.read_csv('enrollment_data.csv')
df
```

```python
plt.xlabel('Year')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 Year')

plt.plot(df['Year'], df['Programming'])
```

```python
plt.xlabel('Year')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 Year')

plt.plot(df['Year'], df['Programming'], label = 'Programming')
plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing')

plt.legend()
```

# Customizing Line plot ( color,linestyle,marker) and Grid

```python
plt.xlabel('Year')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 Year')

plt.plot(df['Year'], df['Programming'], label = 'Programming', color = '#11b86f', linewidth = 2.5, marker = 'o')
plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing', color = 'red', linestyle= 'dashed')

plt.legend()
```

```python
# grid

plt.xlabel('Year')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 Year')

plt.plot(df['Year'], df['Programming'], label = 'Programming', color = '#11b86f', linewidth = 2.5, marker = 'o')
plt.plot(df['Year'], df['Digital Marketing'], label = 'Digital Marketing', color = 'red', linestyle= 'dashed')

plt.legend()
plt.grid()

plt.show()
```

# scatters plots

```python
students = pd.read_csv('student_IQdata.csv')
students
```

```python
# Relation between Study Hours and IQ Score --> positive correlation

plt.xlabel('Study Hours')
plt.ylabel('IQ Score')
plt.title('Relation between Study Hours and IQ Score')

plt.scatter(students['Study_Hour'], students['IQ_Score'])
```

```python
# Relation between Shoe Size and IQ Score --> no correlation

plt.xlabel('Shoe Size')
plt.ylabel('IQ Score')
plt.title('Relation between Shoe Size and IQ Score')

plt.scatter(students['Shoe_Size'], students['IQ_Score'])
```

```python
# Relation between Chilling Hours and IQ Score --> negetive correlation

plt.xlabel('Chilling Hours')
plt.ylabel('IQ Score')
plt.title('Relation between Chilling Hours and IQ Score')

plt.scatter(students['Chilling_Hours'], students['IQ_Score'], marker = 'x', color = 'red')
```

# bar chat

```python
data = {
    'Student': ['B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [2, 5, 3, 7, 4, 8, 1],
    'Attendance': [60, 88, 70, 95, 80, 98, 55],
    'Grade': ['Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# gender wise

gender_group = df.groupby('Gender').size()
gender_group
```

```python
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('bar Chat of Gender among Students')

plt.bar(gender_group.index, gender_group.values, color=['blue', 'red'], edgecolor='black')
```

```python
# bar chat on real data

df = pd.read_csv('student_data.csv')
df
```

```python
# CompletionStatus

status = df.groupby('CompletionStatus').size()

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chat of CompletionStatus among the  Students')

plt.bar(status.index, status.values, color=['green', 'blue', 'red'], edgecolor='black')
```

```python
# location

status = df.groupby('Location').size()

plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Bar Chat of Location among the  Students')

plt.bar(status.index, status.values,color=['blue'], edgecolor='black')
```

# pie chart

```python
status = df.groupby('CompletionStatus').size()

plt.pie(status, labels= status.index, autopct='%1.1f%%', colors=['green', 'blue', 'red'], explode=(0.1, 0, 0), shadow=True)
plt.show()
```
