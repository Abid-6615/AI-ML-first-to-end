# File Read

### File open

```python
file = open("./sample_data/sample.txt", "r")

contant = file.read()

print(contant)

file.close()

print (file.closed)
```

### File open 'with'

```python
with open("./sample_data/sample.txt", "r") as file:
    contant = file.read()
    print(contant)

print(file.closed)
```

# File Write

```python
# Creating a new file and writing there

with open("./sample_data/test.txt", "w") as file:
    file.write("Hello test file\n")
    file.write("How is learning going on?\n")
```

```python
# Writing on a existing file
# its overwrites the file

with open("./sample_data/test.txt", "w") as file:
    file.write("Second attempt\n")
    file.write("Learning is going fast\n")
```

```python
# No overwring just append the text
# append mode

with open("./sample_data/test.txt", "a") as file:
    file.write("The learning of AI/ML is fun\n")
    file.write("I'm enjoying the classes")
```

# File Pointer

```python
# tell (the cursor pointer)

with open("./sample_data/sample.txt", "r") as file:
    print(file.tell())

    print(file.read(5))

    print(file.tell())

    print(file.read())
```

```python
# seek (move the cursor ppointer)
with open("./sample_data/sample.txt", "r") as file:
    print(file.tell())

    print(file.read(5))

    print(file.tell())

    file.seek(0)
    print(file.tell())
```
