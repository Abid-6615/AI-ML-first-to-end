# Variable Type Custing

```python
a = input()
a = int(a)
print(f"Value --> {a} \nType --> {type(a)}")
```

```python
row_numbers = input()
numbers = row_numbers.split()

num_1 = int(numbers[0])
num_2 = int(numbers[1])
num_3 = int(numbers[2])

print(f"Value --> {numbers}, \nType --> {type(numbers)}")

print(f"\nFirst value --> {num_1}, Type --> {type(num_1)} \nSecond value --> {num_2}, Type --> {type(num_2)} \nThird value --> {num_3}, Type --> {type(num_3)}")
```

### If Else Condition

```python
raining = True
muddy = True

if raining:
    print("Don't forget to take umbrella")
    if muddy:
        print("Were your rain boots")
    else:
        print("Were your regular shoes")
else:
    print("Enjoy the sunshine")
```

### Loop Condition

```python
for i in range(1, 6):
    for j in range(1, 4):
        print(f"{i} * {j} = {i*j}")
```
