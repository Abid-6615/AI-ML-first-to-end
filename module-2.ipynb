# Structure of a Function

```python
def adder(a, b):
    """
    add two numbers
    a --> int
    b --> int
    """
    sum = a + b
    return sum
```

```python
x = adder(10, 30) * 5

print(f"The sum is --> {x}")
```

# Different Type of Arguments

```python
def adder(a, b, c):
    print(f"{a}, {b}, {c}")

adder(10, 30, 20)
```

### Positional Argument

```python
def adder(*args):
    for i in args:
        print(i)

adder(10, 20, 30, 40, 50)
```

### Keyword Argument

```python
def adder(a, b, c):
    print(f"{a}, {b}, {c}")

adder(a=10, c=30, b=20)
```

```python
def adder(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} --> {value}")

adder(a=10, b=12, c=14, d=16, e=18, f=20)
```

# Return type of a function

```python
def adder(a, b):
    sum = a + b
    return sum

print(type(adder(10, 20)))
```

```python
def adder(a, b):
    sum = a + b
    return sum

print(type(adder(10.5, 20.7)))
```

```python
def adder(a, b):
    return[a, b]

print(type(adder(10.3, 20.7)))

# unzip

x, y = adder(10.3, 20.7)

print(y)
```

# Iterator

```python
s = {10, 20, 30, 40, 50}

s_iter = iter(s)
print(next(s_iter))
print(next(s_iter))
print(next(s_iter))
print(next(s_iter))
print(next(s_iter))

print("-" * 3)

for i in s:
    print(i)
```

# Generator

```python
lst = [x for x in range(500)]

def data_loader(chunk_size, lst):
    for i in range(0, chunk_size):
        yield lst[i: i+chunk_size]

x = data_loader(5, lst)
print(next(x))
print(next(x))
```

# Lambda Function

```python
import pandas as pd

df = pd.DataFrame({'temperature (F)': [32, 212, 98.6, 68]})
df

df['temperature (C)'] = df['temperature (F)'].apply(lambda x: (x - 32) * 5/9)
print(df)
```

# Map Function

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]})
df

df['A'] = df['A'].map(lambda x: x** 0.5)

print(df)
```

# Apply Function

```python
import pandas as pd

df = pd.DataFrame({'total_spent': [120, 600, 450, 800]})

# Creating a 'High Value' feature on the fly
df['is_high_value'] = df['total_spent'].apply(lambda x: 1 if x > 500 else 0)
df
```
