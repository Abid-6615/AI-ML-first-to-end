# Object Oritented Programming (OOP)

## Class and Object Implementation

```python
# base class / parent class
class Phone():

    # Constructor
    def __init__ (self, brand, model, camera, battery, btry_percentage=100):
        self.brand = brand
        self.model = model
        self.camera = camera
        self.battery = battery
        self.btry_percentage = btry_percentage

    # Mathod
    def charge(self, minute):
        print(f"Charging for {minute} minutes...")
        
    def capture(self, photo):
        if (self.btry_percentage) <= photo:
            print("No charge available for photo capture.")
        else:
            self.btry_percentage -= photo
            print(f"Photo captured on {self.model}")


class Cooling():
    def __init__(self, cool):
        self.cool = cool
        print("Cooling system is 'ON'")
```

## Object Define

```python
iphone = Phone("iPhone", "17 Pro Max", 50, 4500)
xiaomi = Phone("Xiaomi", "Note 13 Pro", 50, 4500)
samsung = Phone("Samsung", "S23 Ultra", 100, 5000)

print(f"Brand name --> {iphone.brand} \nModel name --> {iphone.model} \nBattery --> {iphone.battery} MHz \nCamera --> {iphone.camera} MP")

iphone.capture(35)
```

## Single Inharitance

```python
# child class
class SmartPhone(Phone):
    # init mathod
    def __init__(self,brand, model, camera, battery, processor):
        super().__init__(brand, model, camera, battery)
        self.processor = processor

    def charge(self, minute):
        print("Fast charging is connected...")
        super().charge(minute)
```

```python
vivo = SmartPhone("Vivo", "X2000", 100, 4500, "SnapDragon")
print(vivo.processor)
vivo.charge(10)
```

## Multiple Inharitance

```python
class ProPhone(SmartPhone, Cooling):
    def __init__(self, brand, model, camera, battery, processor, cool):
        SmartPhone.__init__(self, brand, model, camera, battery, processor)
        Cooling.__init__(self, cool)
```

```python
pixel = ProPhone("Pixel", "11 Pro", 120, 5500, "H-750", "Hydrogen")

print(pixel.brand)
print(pixel.model)
print(pixel.camera)
print(pixel.cool)
```

## Polymorphism

```python
class Camera():
    def __init__(self, name):
        self.name = name

    # Mathod
    def capture(self):
        print("Photo is catured")
```

```python
class Phone(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resultion = resulation

    # Mathod Overwrite
    def capture(self):
        print("Photo clicked by phone")
        
class DSLR(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resultion = resulation

    def capture(self):
        print("Photo clicked by DSRL")
        
class Drone(Camera):
    def __init__(self, name, resulation):
        super().__init__(name)
        self.resultion = resulation

    def capture(self):
        print("Photo clicked by Drone")
    
```

```python
phone = Phone("Phone", 50)
dslr = DSLR("DSLR", 200)
drone = Drone("Drone", 150)

phone.capture()
dslr.capture()
drone.capture()
```

## Encapsulation

```python
class Mobile():
    def __init__(self, name, model, imei):
        self.__name = name      # Private
        self.__model = model    # Private
        self.__imei = imei      # Private

    def imei_getter(self):      # Get access
        return self.__imei
        
    def model_getter(self):     # Get access
        return self.__model
        
    def name_getter(self):      # Get access
        return self.__name

    def name_setter(self, name):     # Set value
        self.__name = name
        
        
```

```python
iphone = Mobile("IPhone", "17 Pro", "1xx5904045k")

print(iphone.name_getter())
print(iphone.model_getter())
print(iphone.imei_getter())

print(iphone.name_getter())

iphone.name_setter("iphone")
print(iphone.name_getter())
```

## Abstruction
