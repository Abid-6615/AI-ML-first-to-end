# Handling Duplicates Values

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "David", "Bob"],
    "City": ["New York", "London", "Paris", "New York", "Tokyo", "London"],
    "Score": [85, 90, 78, 70, 95, 90]
}

df = pd.DataFrame(data)
df
```

```python
df.duplicated()
```

```python
# permanently delet duplicate values

df.drop_duplicates(inplace=True)

df
```

```python
# name and city basis delet duplicate values

df.drop_duplicates(subset=['Name', 'City'], keep='last')
```

# Apply Function in DF

```python
df1
```

```python
# min & max scaled value
mn = df1['Total Marks'].min()
mx = df1['Total Marks'].max()

df1['Scaled Marks'] = df1['Total Marks'].apply(lambda x : (x - mn)/(mx-mn))
df1
```

```python
# custom function

def grading_system(marks):
    if marks >= 260:
        return 'A+'
    elif marks >= 250:
        return 'A'
    else:
        return 'A-'

df1['Grade'] = df1['Total Marks'].apply(grading_system)
```

```python
df1
```

```python
def marking_system(df):
    a = df['Data Structure Marks'] * 2
    b = df['Algorithm Marks'] * 3
    c = df['Python Marks'] * 4

    return a+b+c

df1['Exeptional Marks'] = df1.apply(marking_system, axis=1)
```

# group by in pandas

```python
df2 = pd.read_csv('Student_Completed_Data.csv')
df2
```

```python
import numpy as np

df2.loc[df2['FullName'] == 'Alif Rahman', 'Total Marks'] = np.nan
df2.loc[df2['FullName'] == 'Sadia Chowdhury', 'Total Marks'] = np.nan
df2.loc[df2['FullName'] == 'Afsana Mimi', 'Total Marks'] = np.nan
```

```python
df2
```

```python
 instructor_group = df2.groupby('Instructor')

 instructor_group
```

```python
for instructor, df_ins in instructor_group:
    print(instructor)
    display(df_ins)
```

```python
instructor_group.min()
```

```python
mean_marks = instructor_group['Total Marks'].mean()
mean_marks
```

```python
df2['Total Marks'] = df2['Total Marks'].fillna(df2['Instructor'].map(mean_marks))

df2
```
