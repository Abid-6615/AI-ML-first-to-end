```python
import numpy as np
import pandas as pd

np.random.seed(24)

num_employees = 50  # you can increase this if needed

# Employee data
emp_ids = np.arange(101, 101 + num_employees)
emp_ages = np.random.randint(20, 55, num_employees)
emp_income = np.random.randint(35000, 95000, num_employees)

# Fixed set of roles (repetition allowed)
roles_list = [
    "HR Manager",
    "Software Developer",
    "Accountant",
    "Marketing Executive",
    "Sales Executive",
    "Backend Developer",
    "HR Executive",
    "Financial Analyst",
    "Digital Marketer",
    "Operations Manager"
]

emp_roles = np.random.choice(roles_list, num_employees)

# Create DataFrame
data = pd.DataFrame({
    "emp_id": emp_ids,
    "emp_age": emp_ages,
    "emp_salary": emp_income,
    "role": emp_roles
})

# Save dataset
data.to_csv("staff_data.csv", index=False)
```

# Problem solving using Numpy

# 1-Q1

```python
"""
Use NumPy arrays from the emp_ages and emp_salary columns.
Print both arrays and display:
data type
number of dimensions
shape
total number of elements
"""

import numpy as np
data = np.genfromtxt('staff_data.csv', delimiter=',', skip_header=1)

emp_age = data[:, 1]
emp_salary = data[:, 2]

print(f"Employee Ages: --> {emp_age}")
print(f"Employee Salaries --> {emp_salary}")

print(f"\nData Type of emp_ages --> {emp_ages.dtype} \nData Type of emp_salaries --> {emp_salary.dtype}")
print(f"\nnumber of dimensions of emp_ages --> {emp_ages.ndim} \nnumber of dimensions of emp_salaries --> {emp_salary.ndim}")
print(f"\nshape of emp_ages --> {emp_ages.shape} \nshape of emp_salaries --> {emp_salary.shape}")
print(f"\ntotal number of elements of emp_ages --> {emp_ages.size} \ntotal number of elements of emp_salaries --> {emp_salary.size}")
```

# 1-Q2

```python
"""
Using NumPy functions:
Find the maximum and minimum salary
Calculate the average salary
Calculate the average age
Print all results.
"""

max_salary = np.max(emp_salary)
min_salary = np.min(emp_salary)
avg_salary = np.average(emp_salary)
avg_age = np.average(emp_age)

print(f"The height salary is --> {max_salary}")
print(f"The lowest dalary is --> {min_salary}")
print(f"The avarage salary is --> {avg_salary}")
print(f"The avarage age is --> {avg_age}")
```

# 1-Q3

```python
"""
From the emp_ages array:
Extract employees who:
are aged between 25 and 40,
 AND
Emp_salary : have salary greater than 50,000
Then :
Print the filtered values
Count how many employees satisfy this condition
"""

filtered = np.where((emp_age >= 25) & (emp_age <= 40) & (emp_salary > 50000))

filtered_ids = emp_ids[filtered]
filtered_ages = emp_age[filtered]
filtered_salaries = emp_salary[filtered]

for i in range(len(filtered_ids)):
  print(f"ID --> {filtered_ids[i]} |  Age --> {filtered_ages[i]} |  Salary --> {filtered_salaries[i]}")

count = len(filtered_ids)
print(f"\nTotal employees satisfy this condition are --> {count}")
```

# 1-Q4

```python
"""
Create a new NumPy array where each employee’s age is adjusted based on:
Add 4 years if age is less than the average age
Add 2 years if age is greater than or equal to the average age
Do NOT modify the original array.

Print:
original array
updated array
average age used for condition
"""
avg_age = np.mean(emp_age)

updated_data = np.where(emp_age < avg_age, emp_age+4, emp_age+2)

print(f"Original array --> {emp_age}")
print(f"\nUpdated array --> {updated_data}")
print(f"\nAverage age used for condition is --> {avg_age}")
```

# 1-Q5

```python
"""
Using NumPy:
Assume all salaries of Tech roles increase by 8%
Calculate the new total salary expense
Find the overall increase in expense after the raise

Print both old and new totals along with the difference.
"""

selected_roles = ["Software Developer", "Backend Developer", "Digital Marketer"]

new_salary = emp_salary.copy().astype(float)

tech_mask = np.isin(emp_roles, selected_roles)
new_salary[tech_mask] *= 1.08

old_salary = np.sum(emp_salary)
new_salary = np.sum(new_salary)
difference = new_salary - old_salary

print(f"The old total salary expense is --> {old_salary}")
print(f"The new total salary expense is --> {new_salary}")
print(f"Overall increase in expence is --> {difference}")
```

```python
"""
Load the staff_data.csv file into a Pandas DataFrame.
Display the first 3 rows
Show dataset information using .info()
Display summary statistics using .describe()
"""


import pandas as pd
df = pd.read_csv('staff_data.csv')

print("--- First 3 Rows ---")
print(df.head(3))

print("\n--- Dataset Information ---")
df.info()

print("\n--- Summary Statistics ---")
print(df.describe())
```

# 2-Q7

```python
"""
Filter the dataset to include only:
emp_id, emp_age, role
Then:
Select employees whose age is between 25 and 40
Sort them by emp_id
Display the last 3 rows of the result
"""

filtered_df = df[['emp_id', 'emp_age', 'role']]

filtered_df = filtered_df[(filtered_df['emp_age'] >= 25) & (filtered_df['emp_age'] <= 40)]

filtered_df = filtered_df.sort_values('emp_id')

last_3_rows = filtered_df.tail(3)
print(last_3_rows)
```

```python
"""
Sort the DataFrame by:
emp_age (ascending)
and then by emp_salary (descending)
Display the top 5 employees after sorting, including:
emp_id
role
emp_salary
"""

df_sort = df.sort_values(['emp_age', 'emp_salary'], ascending=[1, 0])
print("--- The top 5 employees after sorting ---")
df_sort.head(5)
```

# 2-Q10

```python
"""
Using .loc:
Increase salary by 12% for "Software Developer" and "Backend Developer"
Increase salary by 5% for all other roles
Then:
Print the dataframe that includes role , previous_salary , new_salary , increase_of_salary
"""

df_inc = df.copy()
df_inc['previous_salary'] = df_inc['emp_salary']

dev_filter = df_inc['role'].isin(['Software Developer', 'Backend Developer'])

df_inc.loc[dev_filter, 'emp_salary'] = df_inc['emp_salary'] * 1.12
df_inc.loc[~dev_filter, 'emp_salary'] = df_inc['emp_salary'] * 1.05

df_inc['new_salary'] = df_inc['emp_salary']
df_inc['increase_of_salary'] = df_inc['new_salary'] - df_inc['previous_salary']

print(df_inc[['role', 'previous_salary', 'new_salary', 'increase_of_salary']])
```
